```python
import matplotlib.pyplot as plt

def plotfig(data, title, label):
    x = [i for i in range(len(data[0]))]
    
    for d,l in zip(data,label):
        plt.plot(x, d, 'o-',label=l)
        for a, b in zip(x, d):
            plt.text(a, b + 0.02*d[0], '%.2f' % b, ha='center', va='bottom', fontsize=7)
        
    plt.xticks(x, x)
    plt.xlabel("generation")
    plt.title(title)
    plt.legend()
    plt.savefig(title+'.png')
    plt.show()
```

```python
from GenAlg import GenAlg
from Fitness import *
from Crossover import *
from Evaluation import *

l = 40
N = 200

fitness = counting_ones
crossover = two_x

GA = GenAlg(l, N, crossover, fitness)

print(GA.run())

print(GA.generation)


plotfig([GA.prop], 'proportion of bits-1', ['prop'])

plotfig([GA.correct,GA.error], 'number of correct\error selection',['correct','error'])

num1 = [a[0] for a in GA.schematadata]
num0 = [a[1] for a in GA.schematadata]
sch1 = [a[2] for a in GA.schematadata]
std1 = [a[3] for a in GA.schematadata]
sch0 = [a[4] for a in GA.schematadata]
std0 = [a[5] for a in GA.schematadata]

plotfig([num1, num0], 'number of schemata', ['num 1**', 'num 0**'])

plotfig([sch1, sch0], 'average fitness schemata', ['avg 1**', 'avg 0**'])

plotfig([std1, std0], 'standard deviation schemata', ['std 1**','std 0**'])
```
